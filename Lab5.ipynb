```python
import pandas as pd

df = pd.DataFrame({
    'Total_Price': [45.50, 12.00, 89.99, 25.00],
    'Item_Count': [3, 1, 5, 2],
    'Order_Time': pd.to_datetime(['2023-10-24 18:30:00', '2023-10-24 14:15:00', '2023-10-24 22:45:00', '2023-10-24 09:10:00']),
    'Order_Status': ['Delivered', 'Cancelled', 'In Transit', 'Delivered']
})
df['Average_Item_Price'] = df['Total_Price'] / df['Item_Count']
```

The Average_Item_Price feature distinguishes premium or bulk orders from standard low-cost deliveries. Unusually high average item prices might correlate with a higher probability of cancellations due to customer hesitation or payment issues, providing a stronger signal for Order_Status than total price alone.

```python
df['is_peak_hour_extended'] = df['Order_Time'].dt.hour.apply(lambda x: 1 if 17 <= x <= 23 else 0)
```

Expanding the peak hour window to 17:00–23:00 captures both early dinner and late-night rushes. If classification performance improves, it indicates the original model missed critical operational congestion periods that frequently lead to delayed or cancelled orders.
